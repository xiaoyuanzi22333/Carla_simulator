```python
import numpy as np
import matplotlib.pyplot as plt

root = './output_test_001/2025_01_06_19-59-34'


steer = np.load(root + "/steer_data.npy")
brake = np.load(root + "/brake_data.npy")
throttle = np.load(root + '/throttle_data.npy')
gear = np.load(root + '/gear_data.npy')

print(len(steer))
# print(len(brake))
# print(len(throttle))
# print(len(gear))

x = np.arange(len(steer))


plt.plot(x, steer, label='steer', color='blue', linestyle='--')
plt.plot(x, brake, label='brake', color='green', linestyle='-.')
plt.plot(x, throttle, label='throttle', color='red', linestyle=':')
# plt.plot(x, gear, label='gear', color='purple', linestyle='-.')

# 
plt.legend()

# 
plt.title('Multiple Data Series on One Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# 
plt.show()


```
